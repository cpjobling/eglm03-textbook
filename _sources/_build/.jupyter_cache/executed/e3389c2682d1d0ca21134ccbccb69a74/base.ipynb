```matlab
Kt = 0.5; K = 8;
```

```matlab
clce1 = [1, K*Kt, K];
```

```matlab
Gc1 = tf(K,clce1)
```

```matlab
Kt = 1/2;
z1 = -1/Kt;
Go = zpk(z1,[0, 0],1) % root locus gain initially set to unity
```

```matlab
rlocus(Go),title('Root locus for cascade compensated system')
```

```matlab
Kc = rlocfind(Go,-2+2j)
```

```matlab
D = tf(Kc*[1 -z1],1)
```

```matlab
G=tf(1,[1,0,0])
```

```matlab
Gc2 = feedback(D*G,1)
```

```matlab
[y1,t1]=step(Gc1);
[y2,t2]=step(Gc2);
plot(t1,y1,t2,y2),...
 legend('Velocity fb','P+D'),...
 title('Step response: closed-loop compensated system')
```
