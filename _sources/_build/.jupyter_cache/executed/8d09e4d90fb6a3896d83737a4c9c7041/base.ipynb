```matlab
p = 5; % rad/s
ds = tf([p], [1 p])
```

```matlab
Fs = (20*p)/(2*pi); % Hz
Ts = 1/Fs;
```

```matlab
dz = c2d(ds, Ts, 'zoh')
```

```matlab
step(ds,'-',dz,'--')
```

```matlab
ds = tf(5, [1 5])
bode(ds)
grid
```

```matlab
Fs = 15;
Ts = 1/Fs;
```

```matlab
dsz=c2d(ds, Ts, 'zoh')
```

```matlab
dst = c2d(ds, Ts, 'tustin')
```

```matlab
dsm = c2d(ds, Ts, 'mpz')
```

```matlab
bode(ds,dsz,'*',dst,'+',dsm,'o')
grid
legend('D(jw)','ZOH','Tustin','MPZ')
title('Magnitude and phase of discrete equivalents for D(s) = 5/(s+5) - Fs = 15 Hz')
```
